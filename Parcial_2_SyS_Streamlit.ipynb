```python
%%writefile pages/Punto_1.py
import streamlit as st
import numpy as np
import time

st.set_page_config(page_title="Punto 1", page_icon="üé∂")

st.sidebar.header("Punto 1")

st.markdown("## Punto 1")

st.write("Sea el demodulador en amplitud presentado en la siguiente figura:")

#st.image("https://drive.google.com/uc?export=view&id=1J0x60UsK_QW6WkTn6BoiE1wqNCVCy96W")

st.image("demoduladorpar2sys.png")

st.markdown(r"""
Asumiendo $$ \theta_0 = 0 $$, determine el espectro de Fourier (te√≥rico) en cada una de las etapas del sistema.

Luego, con base en la simulaci√≥n de amplitud del Taller 2 y utilizando 5 segundos de una canci√≥n de YouTube como mensaje, grafique cada una de las etapas principales del proceso de modulaci√≥n y demodulaci√≥n en el tiempo y en la frecuencia (reproduzca el segmento de la canci√≥n en cada etapa).

Nota: Para la etapa de filtrado pasa bajas, realice su implementaci√≥n a partir de la transformada r√°pida de Fourier (FFT).
""")

st.markdown("### R=/")
st.markdown("""
Lo primero que nos piden es determinar el espectro de Fourier en cada etapa del sistema asumiendo $$Œ∏_0=0$$.""")

st.markdown(r"""
Sea nuestra se√±al *"DSB-CS"*:
""")

st.markdown(r"""
$$
A_1 m(t)\cos(2\pi f_0 t)
$$
""")

st.markdown("Sabemos que:")

st.markdown(r"""
$$
\cos(\theta)=\frac{e^{j\theta}+e^{-j\theta}}{2}
$$
""")

st.markdown("Luego:")

st.markdown(r"""
$$
\cos(2\pi f_0 t)=\frac{e^{j2\pi f_0 t}+e^{-j2\pi f_0 t}}{2}
$$
""")

st.markdown("Con lo cual:")

st.markdown(r"""
$$
A_1 m(t)\cos(2\pi f_0 t)
=
\frac{A_1}{2}\left[m(t)e^{j2\pi f_0 t}+m(t)e^{-j2\pi f_0 t}\right]
$$
""")

st.markdown("As√≠, al aplicar Fourier:")

st.markdown(r"""
$$
F\{A_1 m(t)\cos(2\pi f_0 t)\}
=
\frac{A_1}{2}\left[
F\{m(t)e^{j2\pi f_0 t}\}
+
F\{m(t)e^{-j2\pi f_0 t}\}
\right]
$$
""")

st.markdown(r"""
Para aplicar transformada de Fourier, recordemos que:
""")

st.markdown(r"""
$$
F\{x(t)\}=\int_{-\infty}^{\infty}x(t)e^{-jwt}\,dt=X(\omega)
$$
""")

st.markdown("Luego:")

st.markdown(r"""
$$
F\{x(t)e^{\pm j\omega_0 t}\}
=
\int_{-\infty}^{\infty} x(t)e^{-j(\omega\mp\omega_0)t}\,dt
=
X(\omega \mp \omega_0)
$$
""")

st.markdown("Por lo tanto:")

st.markdown(r"""
$$
F\{A_1 m(t)\cos(2\pi f_0 t)\}
=
\frac{A_1}{2}\Big[
M(\omega-2\pi f_0) + M(\omega+2\pi f_0)
\Big]
$$
""")

st.markdown(r"""
Luego del *"mixer"* la se√±al queda:
""")

st.markdown(r"""
$$
A_1 m(t)\cos^2(2\pi f_0 t)
=
\frac{A_1 m(t)}{2}
+
\frac{A_1 m(t)\cos(4\pi f_0 t)}{2}
$$
""")

st.markdown(r"""
Nuevamente:
""")

st.markdown(r"""
$$
\cos(\theta)=\frac{e^{j\theta}+e^{-j\theta}}{2}
$$
""")

st.markdown(r"""
Luego:
""")

st.markdown(r"""
$$
\cos(4\pi f_0 t)=\frac{e^{j4\pi f_0 t}+e^{-j4\pi f_0 t}}{2}
$$
""")

st.markdown(r"""
As√≠, el espectro de Fourier ser√°:
""")

st.markdown(r"""
$$
F\{A_1 m(t)\cos^2(2\pi f_0 t)\}
=
F\left\{
\frac{A_1 m(t)}{2}
+
\frac{A_1 m(t)\cos(4\pi f_0 t)}{2}
\right\}
$$
""")

st.markdown(r"""
$$
=
\frac{A_1}{2}
\left[
F\{m(t)\}
+
\frac{1}{2}F\{m(t)e^{j4\pi f_0 t}\}
+
\frac{1}{2}F\{m(t)e^{-j4\pi f_0 t}\}
\right]
$$
""")

st.markdown(r"""
$$
=
\frac{A_1}{2}
\left[
M(\omega)
+
\frac{M(\omega - 4\pi f_0) + M(\omega + 4\pi f_0)}{2}
\right]
$$
""")

st.markdown(r"""
Luego del *"Lowpass Filter"* la se√±al queda:
""")

st.markdown(r"""
$$
\frac{A_1 m(t)}{2}
$$
""")

st.markdown(r"""
Por lo cual su espectro de Fourier es evidentemente:
""")

st.markdown(r"""
$$
F\left\{\frac{A_1 m(t)}{2}\right\}
=
\frac{A_1 M(\omega)}{2}
$$
""")

st.markdown(r"""
Finalmente, luego del *"Scale Amplitude"*:
""")

st.markdown(r"""
$$
F\{m(t)\} = M(\omega)
$$
""")

st.markdown(r"""
Despu√©s de hacer las transformadas de Fourier a mano, nos piden usar una canci√≥n de YouTube como mensaje y graficar cada una de las etapas principales del proceso de modulaci√≥n y demodulaci√≥n en el tiempo y en la frecuencia.

Para esto, primero vamos a descargar una canci√≥n:
""")

import soundfile as sf

nombre_out = "output.wav"

x, fs = sf.read(nombre_out)

ns = 30  # segundos
x30 = x[:int(fs * ns)]

sf.write("temp_30s.wav", x30, fs)


with open("temp_30s.wav", "rb") as f:
    audio_bytes = f.read()

st.audio(audio_bytes, format="audio/wav")



st.markdown("""
Ya tenemos nuestro mensaje $$m(t)$$ que en Python nombramos como $$x$$,
ahora vamos a decidir los valores de $$A‚ÇÅ$$ y $$f‚ÇÄ$$.

Seg√∫n el ejemplo de modulaci√≥n AM, y seg√∫n he consultado, $$A‚ÇÅ$$ es la amplitud del canal.
En esta ocasi√≥n, por simpleza elegimos:
""")

st.latex(r"A_1 = 1")

st.markdown("""
Sin embargo, $$f‚ÇÄ$$ no se puede elegir tan libremente, ya que es la frecuencia de la portadora, y lo normal es que la frecuencia de la portadora sea mucho mayor que la del mensaje.

En el ejemplo de modulaci√≥n AM se us√≥ como frecuencia del mensaje $$10 Hz$$
y como frecuencia de la portadora $$100 Hz$$.

A pesar de eso, por recomendaci√≥n del profesor, ya que esto es solo con fines educativos usaremos una onda de menor frecuencia para montar nuestra canci√≥n
y no tener que resamplear m√∫ltiples veces.
""")

import streamlit as st
import matplotlib.pyplot as plt
import numpy as np

Fc = 15000  # frecuencia de portadora
ti = 0
Fmuestreo = fs  #frecuencia de muestreo de portadora = frecuencia de muestreo de mi canci√≥n para poder multiplicarlas
tf = 5  #para que grafique 5 segundos

t = np.arange(ti, tf, 1/Fmuestreo)# vector del tiempo

xpro = x.copy() #copiar archivos para procesar
xs = xpro[int(ti):int(tf*Fmuestreo),:]
c= np.cos(2*np.pi*Fc*t)
fig, ax = plt.subplots() #hay que nombrarla pa que streamlit la pueda mostrar

ax.plot(t, c, label='portadora')
ax.plot(t, xs, label='mensaje')

ax.set_xlabel("Tiempo (s)")
ax.set_ylabel("Amplitud")
ax.set_title("Mensaje y portadora")
ax.legend()

st.pyplot(fig)

st.markdown("Mientras tanto la gr√°fica en frecuencia queda:")

fig2, ax2 = plt.subplots()

Xw = np.fft.rfft(xs,axis=0) # axis=0 permite aplicar fft por cada columna de xpro
vf = np.fft.rfftfreq(np.size(xs,0),1/Fmuestreo) #se crea el vector de frecuencias
cw = np.fft.rfft(c, axis=0)
vfc = np.fft.rfftfreq(np.size(c,0),1/Fmuestreo) #se crea el vector de frecuencias
ax2.plot(vf,abs(Xw), label='Canci√≥n original')#se grafica la magnitud
ax2.plot(vfc,abs(cw), label='Portadora')#se grafica la magnitud
ax2.legend()
ax2.set_title('Espectro audio resampleado')
ax2.set_xlabel('$f[Hz]$',fontsize = 14)
ax2.set_ylabel('$|X[n]|$',fontsize = 14)

st.pyplot(fig2)

st.markdown("Ya que no se ve el espectro de nuestra canci√≥n original, lo voy a pintar a parte:")

fig3, ax3 = plt.subplots()

ax3.plot(vf,abs(Xw), label='Canci√≥n original')#se grafica la magnitud
ax3.legend()
ax3.set_title(r'Espectro audio resampleado')
ax3.set_xlabel(r'$f[Hz]$',fontsize = 14)
ax3.set_ylabel(r'$|X[n]|$',fontsize = 14)

st.pyplot(fig3)

st.markdown("Y vemos que era de esperarse que no se viera nuestra canci√≥n, ya que tiene magnitudes de apenas  $$3000$$ , mientras la portadora llega a  $$10^9$$")

st.markdown("""Ahora vamos a crear con base a estas 2 se√±ales la se√±al *"received DSB-CS signal"*""")

fig4, ax4 = plt.subplots()

xs_mono = np.mean(xs, axis=1) #convertimos de stereo a mono para poder enviar el mensaje
#es la soluci√≥n m√°s simple, y la que usa AM tradicional, ya que sino tendriamos que o enviar 2 ondas una con cada canal o hacer un arreglo m√°s complejo.
received = c*xs_mono
ax4.plot(t,received,label='Received DSB-CS signal')
ax4.legend()

st.pyplot(fig4)

st.markdown("En un principio parece que es la misma canci√≥n que al inicio, pero si la escuchamos:")

ns = 5  # segundos
x5 = received[:int(fs * ns)]

sf.write("temp_5s.wav", x5, fs)


with open("temp_5s.wav", "rb") as f:
    audio_bytes = f.read()

st.audio(audio_bytes, format="audio/wav")

st.markdown("Es un pitido feo en el que apenas se escucha nuestra canci√≥n, el cambio es a√∫n mas notorio al investigar la frecuencia:")

fig5, ax5 = plt.subplots()

Xwr = np.fft.rfft(received,axis=0) # axis=0 permite aplicar fft por cada columna de xpro
vfr = np.fft.rfftfreq(np.size(received,0),1/Fmuestreo) #se crea el vector de frecuencias
ax5.plot(vfr,abs(Xwr), label='Received DBS-CS signal')#se grafica la magnitud
ax5.plot(vf,abs(Xw), label='Canci√≥n original')#se grafica la magnitud
ax5.plot(vfc,abs(cw), label='Portadora')#se grafica la magnitud
ax5.legend()
ax5.set_title(r'Espectro audio resampleado')
ax5.set_xlabel(r'$f[Hz]$',fontsize = 14)
ax5.set_ylabel(r'$|X[n]|$',fontsize = 14)
ax5.set_ylim([abs(Xwr).min(), 1.1*abs(Xwr).max()])

st.pyplot(fig5)

st.markdown("""En este caso agregamos un limite a el eje y para que fuese m√°s visible nuestro espectro. Tal como podemos apreciar si hay diferencia entre la canci√≥n original, la portadora y la *"received DBS-CS signal"*. Ahora continuamos con la siguiente parte del proceso.""")

st.markdown("Luego de recibir esta se√±al, est√° pasa por el mixer y se convierte en:")

st.markdown("""$$A_1 m(t) cos^2 (2* \pi *f_0 *t)$$""")

st.markdown("Vamos a simular ese proceso:")

fig6, ax6 = plt.subplots()

mixer = received * np.cos(2*np.pi*Fc*t)
ax6.plot(t,mixer,label='Post Mixer signal')
ax6.legend()

st.pyplot(fig6)

x6 = mixer[:int(fs * ns)]

sf.write("temp_5s.wav", x6, fs)


with open("temp_5s.wav", "rb") as f:
    audio_bytes = f.read()

st.audio(audio_bytes, format="audio/wav")

st.markdown("Nuevamente vamos a investigar su frecuencia para ver que pasa:")

fig7, ax7 = plt.subplots()

Xwmx = np.fft.rfft(mixer,axis=0) # axis=0 permite aplicar fft por cada columna de xpro
vfmx = np.fft.rfftfreq(np.size(mixer,0),1/Fmuestreo) #se crea el vector de frecuencias
ax7.plot(vfr,abs(Xwr), label='Received DBS-CS signal')#se grafica la magnitud
ax7.plot(vf,abs(Xw), label='Canci√≥n original')#se grafica la magnitud
ax7.plot(vfc,abs(cw), label='Portadora')#se grafica la magnitud
ax7.plot(vfmx,abs(Xwmx), label='Post Mixer')#se grafica la magnitud
ax7.legend(loc='upper right') #a partir de aqui hay que especificar que coloque la leyenda en la parte
#superior derecha, ya que si no la muestra en el centro de la gr√°fica
ax7.set_title(r'Espectro audio resampleado')
ax7.set_xlabel(r'$f[Hz]$',fontsize = 14)
ax7.set_ylabel(r'$|X[n]|$',fontsize = 14)
ax7.set_ylim([abs(Xw).min(), 1.1*abs(Xwmx).max()])

st.pyplot(fig7)

st.markdown("""En esta gr√°fica del espectro es posible ver que luego de pasar por el mixer quedan 2 espectros principales, si hicieramos la gr√°fica sin el valor absoluto posiblemente nos mostraria 3, ya que quedo el mensaje original, multiplicado por un escalar y sumado a una se√±al de una frecuencia a√∫n mayor, lo que es:""")

st.markdown(r"""$$\frac{A_1 m(t)}{2} + \frac{A_1 m(t)}{2}cos(4 \pi f_0 t)$$""")

st.markdown(r"""Ahora tenemos que implementar un filtro pasa bajas que solo deje pasar a nuestro mensaje multiplicado por un escalar $$\frac{A_1}{2}$$, para este fin reciclaremos un c√≥digo usado durante el Taller 2:""")

fig8, ax8 = plt.subplots()

#filtrar espectro
Xwpb = Xwmx.copy() #copiamos el post mixer a un Xwpb
f1 = 0 #frecuencia en Hz corte 1 frecuencia min
f2 = (fs/4)*1.1 #frecuencia en Hz corte 2 frecuencia max, coloque (fs/4)*1.1, pues es el pasabajas que mejor se adapta, si la cancion estaba a 48000 Hz es igual a 13200, poco antes de la frecuencia del ruido, si la cancion esta a 44100 entonces es aproximadamente 12150 que es tambien poco antes del espectro del ruido, se adapta bien pues luego del mixer lo unico que queda es el mensaje original multiplicada por un escalar sumado al mensaje original modulado a una frecuencia aun m√°s alta que los originales 15000 Hz de la onda que se uso como carry, y esa onda modulada a m√°s alta frecuencia es el ruido que se escucha
#Normalmente se usa una frecuencia de corte ligeramente mayor a la frecuencia m√°xima de mi canci√≥n, por ejemplo, (fs/2)*1.1, pero en esta ocasi√≥n no es posible ya que si implementamos un filtro pasa bajas a dicha frecuencia seguiriamos escuchando el ruido seg√∫n lo que podemos ver en la gr√°fica, por eso la modulaci√≥n AM se realiza a una frecuencia mucho m√°s alta que la del mensaje, en este caso con el filtro implementado se pierden agudos pero no se va a escuchar el ruido.
#El escalar no va a afectar a nuestra frecuencia
ind = ~((vf > f1) & (vf < f2)) #frecuencias eliminar-> recueder que ~ actua como negaci√≥n
Xwpb[ind] = 0
ax8.plot(vfr,abs(Xwr), label='Received DBS-CS signal')#se grafica la magnitud
ax8.plot(vf,abs(Xw), label='Canci√≥n original')#se grafica la magnitud
ax8.plot(vfc,abs(cw), label='Portadora')#se grafica la magnitud
ax8.plot(vfmx,abs(Xwmx), label='Post Mixer')#se grafica la magnitud
ax8.plot(vf,abs(Xwpb), label='Post pasa baja')
ax8.legend(loc='upper right')
ax8.set_title(r'Espectro audio resampleado')
ax8.set_xlabel(r'$f[Hz]$',fontsize = 14)
ax8.set_ylabel(r'$|X[n]|$',fontsize = 14)
ax8.set_ylim([abs(Xw).min(), 1.1*abs(Xwpb).max()])

st.pyplot(fig8)

xpb = np.fft.irfft(Xwpb,axis=0) #fft inversa

x8 = xpb[:int(fs * ns)]

sf.write("temp_5s.wav", x8, fs)


with open("temp_5s.wav", "rb") as f:
    audio_bytes = f.read()

st.audio(audio_bytes, format="audio/wav")

st.markdown("Lo que gr√°ficamente luce:")

fig9, ax9 = plt.subplots()

ax9.plot(t,xpb,label='Post Pasa bajas')
ax9.legend()

st.pyplot(fig9)

st.markdown(r"""Esta es practicamente nuestra canci√≥n original, se supone que solo ha sido multiplicada por un escalar $$\frac{A_1}{2}$$, asi que volvemos pasamos a la etapa de *"Scale amplitude"*, donde multplicamos por $$\frac{2}{A_1}$$:""")

fig10, ax10 = plt.subplots()

xsa = xpb*2 #x scale amplitude es igual a x pasa bajas por 2 sobre la amplitud de la portadora
ax10.plot(t,xsa,label='Post Scale amplitude')
ax10.legend()

st.pyplot(fig10)

st.markdown("Si lo comparamos con la canci√≥n original, veremos que es el mismo mensaje inicial, con las mismas amplitudes, solo que en Mono y no en Stereo, si queremos lo podemos ver en el dominio de la frecuencia:")

fig11, ax11 = plt.subplots()

Xwsa = np.fft.rfft(xsa,axis=0) # axis=0 permite aplicar fft por cada columna de xpro
vfsa = np.fft.rfftfreq(np.size(xsa,0),1/Fmuestreo) #se crea el vector de frecuencias
ax11.plot(vfr,abs(Xwr), label='Received DBS-CS signal')#se grafica la magnitud
ax11.plot(vf,abs(Xw), label='Canci√≥n original')#se grafica la magnitud
ax11.plot(vfc,abs(cw), label='Portadora')#se grafica la magnitud
ax11.plot(vfmx,abs(Xwmx), label='Post Mixer')#se grafica la magnitud
ax11.plot(vf,abs(Xwpb), label='Post pasa baja')
ax11.plot(vfsa,abs(Xwsa), label='Post Scale amplitude')#se grafica la magnitud
ax11.legend(loc='upper right') #a partir de aqui hay que especificar que coloque la leyenda en la parte
#superior derecha, ya que si no la muestra en el centro de la gr√°fica
ax11.set_title(r'Espectro audio resampleado')
ax11.set_xlabel(r'$f[Hz]$',fontsize = 14)
ax11.set_ylabel(r'$|X[n]|$',fontsize = 14)
ax11.set_ylim([abs(Xw).min(), 1.1*abs(Xwsa).max()])

st.pyplot(fig11)

st.markdown("Finalmente escuchamos devuelta nuestra canci√≥n original en mono:")

x11 = xsa[:int(fs * ns)]

sf.write("temp_5s.wav", x11, fs)


with open("temp_5s.wav", "rb") as f:
    audio_bytes = f.read()

st.audio(audio_bytes, format="audio/wav")
```

```python
%%writefile pages/Punto_2.py
import streamlit as st
import numpy as np
import time

st.set_page_config(page_title="Punto 2", page_icon="‚ö°")

st.sidebar.header("Punto 2")

st.markdown("## Punto 2")

st.write("""Encuentre la funci√≥n de transferencia que caracteriza el sistema masa, resorte, amortiguador, presentado en la siguiente Figura (asuma condiciones iniciales cero):""")

#st.image("https://drive.google.com/uc?export=view&id=1J0x60UsK_QW6WkTn6BoiE1wqNCVCy96W")

st.image("sistemaresorte.png")

st.markdown("Posteriormente, encuentre el sistema equivalente del modelo masa, resorte, amortiguador, a partir del siguiente circuito el√©ctrico:")

st.image("circuitoRLC.png")

st.markdown(r"""Finalmente proponga unos valores de $$m$$, $$k$$ y $$c$$ y sus equivalentes $$R$$, $$L$$ y $$C$$, para simular un sistema subamortiguado, sobreamortiguado y de amortiguamiento cr√≠tico (determine el factor de amortiguamiento, la frecuencia natural amortiguada, la frecuencia natural no amortiguada, el tiempo pico, tiempo de levantamiento y el tiempo de establecimiento en cada caso). Para cada caso, grafique el diagrama de polos y ceros, el diagrama de Bode, la respuesta impulso, respuesta escal√≥n y respuesta rampa. Repita el proceso para modo lazo cerrado.""")

st.markdown(r"""### R=/
Lo primero que nos piden es hallar la funci√≥n de transferencia que caracteriza el sistema masa, resorte, amortiguador.

Seg√∫n lo aprendido en el cuaderno de Pendulo El√°stico las propiedades mec√°nicas del p√©ndulo el√°stico amortiguado se analizan utilizando la teoria de los SLIT y su configuraci√≥n mec√°nica es;

- Un cuerpo rigido de masa $m$ montado sobre un resorte de rigidez $k$.
- Un amortiguador con coeficiente de amortiguamiento $c$ paralelo al resorte para modelar la fricci√≥n del sistema
- Se asume que el movimiento de la masa a lo largo del tiempo se restringe al eje vertical y.
- La masa no se mueve para $$t<0$$.
- Condiciones iniciales $$y(t)=0$$ y $$\frac{dy(t)}{dt} =0$$ para $$t<0$$

## Ecuaci√≥n diferencial:

La ecuaci√≥n diferencial del sistema mec√°nico se obtiene al considerar el equilibrio de fuerzas ejercidas sobre la masa:""")

st.markdown(r"$$F_E(t) = F_S(t) + F_F(t) + F_I(t)$$")

st.markdown(r"""Donde $$F_E(t)$$ denota una fuerza externa que actua sobre la masa.

La fuerza $$F_S(t)$$ inducida por el resorte esta dada por la ley de Hooke:""")

st.markdown(r"$$F_S(t) = k y(t)$$")

st.markdown(r"""Este $$y(t)$$  es el desplazamiento de la masa con respecto al tiempo.

La fuerza de fricci√≥n $$F_F(t)$$ inducida por el amortiguador se modela de manera proporcional a la velocidad de la masa:

Si $$y(t)$$ es el desplazamiento en funci√≥n del tiempo, la velocidad es su primera derivada $$\frac{dy(t)}{dt}$$ con lo cual nos queda:""")

st.markdown(r"$$F_F(t)= c\frac{dy(t)}{dt}$$")

st.markdown(r"Ahora solo falta la fuerza de inercia $$F_I$$ que debida a la aceleraci√≥n se la masa se modela como:")

st.markdown(r"$$F_I(t)=m\frac{d^2y(t)}{dt^2}$$")

st.markdown(r"""Ya que la aceleraci√≥n es la segunda derivada de la posici√≥n y la primera derivada de la velocidad.

Finalmente si despejamos esto en la expresi√≥n inicial obtenemos:

$$
m\frac{d^2y(t)}{dt^2}+c\frac{dy(t)}{dt}+ky(t)=F_E(t)
$$""")

st.markdown(r"""Esta es la ecuaci√≥n diferencial que describe el desplazamiento del p√©ndulo elastico amortiguado como consecuencia de la fuerza externa $$F_E(t)$$, que puede interpretarse como un SLIT para el cual $$x(t)=F_E(t)$$ es la se√±al de entrada y el desplazamiento de la masa $$y(t)$$ es la se√±al de salida.

Entonces la ecuaci√≥n quedo:

$$
m\frac{d^2y(t)}{dt^2}+c\frac{dy(t)}{dt}+ky(t)=x(t)
$$

Sabemos que la funci√≥n de transferencia viene dada por:

$$
H(s)=\frac{Y(s)}{X(s)}
$$

O tambi√©n:

$$
H(s)=\frac{Y(s)}{X(s)}=\frac{Œ£^M_{m=0}a_ms^M}{Œ£^N_{n=0} b_ns^N}
$$

En esta definici√≥n el $a_m$ es el coeficiente m-√©simo que acompa√±an a a la derivada n√∫mero $m$, sabiendo que el SLIT puede tambi√©n escribirse como:

$$
\Sigma^M_{m=0}a_m\frac{d^m}{dt^m}x(t)=\Sigma^N_{n=0}b_n\frac{d^n}{dt^n}y(t)
$$

Lo que en nuestro caso luce como:

$$
x(t)=ky(t)+c\frac{dy(t)}{dt}+m\frac{d^2y(t)}{dt^2}
$$

Con lo que resulta evidente que:

$$
H(s)=\frac{Y(s)}{X(s)}=\frac{Œ£^M_{m=0}a_ms^M}{Œ£^N_{n=0} b_ns^N} = \frac{1s^0}{ks^0+cs^1+ms^2}
$$

O lo que es lo mismo:

$$
H(s)=\frac{1}{k+cs+ms^2}
$$
""")

st.markdown(r"""Lo siguiente que se nos pide es hallar el sitema equivalente a partir de un circuito el√©ctrico. Para empezar hay que plantear ecuaciones para el circuito:

Analizando por nodos, si tomamos como tierra al nodo de abajo por costumbre, la ecuaci√≥n del nodo $$Vo(t)$$ es:

$$
i_R(t)+i_C(t)+i_L(t)=0
$$

La corriente en la resistencia es el elemento m√°s sencillo de despejar, viene dado por $$\frac{V}{R}$$, y en este caso la tensi√≥n que pasa por la resistencia no es otra que la de salida $$Vo(t)$$, asi que queda:

$$
i_R(t)=\frac{Vo(t)}{R}
$$

El siguiente elemento m√°s sencillo es la corriente del condesador que viene dada por:

$$
i_C(t)=C\frac{d}{dt}v_C(t)
$$

Donde $$V_C(t)$$ es la tensi√≥n en el condensador, y en este circuito resulta evidente que es $$Vo(t)$$, asi que queda:

$$
i_C(t)=C\frac{d}{dt}Vo(t)
$$

Finalmente, se supone que la tensi√≥n en el inductor es:

$$
v_L(t)=L\frac{d}{dt}i_L(t)
$$

Pero en este caso:

$$
v_L(t)=Vo(t)-Vi(t)
$$

Luego:

$$
Vo(t)-Vi(t)=L\frac{d}{dt}i_L(t)
$$

Aplicando algo de √°lgebra:

$$
\frac{Vo(t)-Vi(t)}{L}=\frac{d}{dt}i_L(t)
$$

Si volvemos a la ecuaci√≥n original y reemplazamos lo que ya teniamos:

$$
i_R(t)+i_C(t)+i_L(t)=0
$$

$$
\frac{Vo(t)}{R}+C\frac{d}{dt}Vo(t)+i_L(t)=0
$$

$$
-\frac{Vo(t)}{R}-C\frac{d}{dt}Vo(t)=i_L(t)
$$

Si derivamos a ambos lados con respecto a t obtenemos:

$$
-\frac{1}{R}\frac{d}{dt}Vo(t)-C\frac{d^2}{dt^2}Vo(t)=\frac{d}{dt}i_L(t)
$$
""")

st.markdown(r"""Pero $$\frac{d}{dt}i_L(t)$$ es:

$$
\frac{Vo(t)-Vi(t)}{L}=\frac{d}{dt}i_L(t)
$$

Luego al reemplazar nos queda:

$$
-\frac{1}{R}\frac{d}{dt}Vo(t)-C\frac{d^2}{dt^2}Vo(t)=\frac{Vo(t)-Vi(t)}{L}
$$

Hacemos algebra y finalmente la ecuaci√≥n queda:

$$
CL\frac{d^2}{dt^2}Vo(t)+\frac{L}{R}\frac{d}{dt}Vo(t)+Vo(t)=Vi(t)
$$""")

st.markdown(r"""Para no restringir el valor de k solo a 1, multiplicamos toda la ecuaci√≥n por $$\frac{1}{L}$$ y nos queda:

$$
C\frac{d^2}{dt^2}Vo(t)+\frac{1}{R}\frac{d}{dt}Vo(t)+\frac{1}{L}Vo(t)=\frac{1}{L}Vi(t)
$$""")

st.markdown(r"""Asi podemos decir que el modelo es equivalente si:

$$
C = m
$$

$$
\frac{1}{R} = c
$$

$$
\frac{1}{L} = k
$$

$$
\frac{1}{L}Vi(t) = x(t) = F_E(t)
$$""")

st.markdown(r"""Con esto en mente, la funci√≥n de transferencia del circuito queda:

$$
H(s)=\frac{\frac{1}{L}}{\frac{1}{L}+\frac{1}{R}s+Cs^2}
$$""")

st.markdown(r"""Lo siguiente que nos piden es dar valor a nuestras variables $$R$$, $$L$$, $$C$$, $$m$$, $$c$$ y $$k$$ para simular un sistema subamortiguado, sobreamortiguado, y de amortiguamiento critico, pero, para no ir dando valores al azar calcularemos el factor de amortiguamiento en ambos casos y daremos valores seg√∫n nuestras necesidades.

El factor de amortiguamiento $$Œæ$$ se calcula, teniendo una funci√≥n de transferencia de la forma:

$$
H(s)=\frac{1}{a_2s^2+a_1s+a_0}
$$

Como:

$$
Œæ=\frac{a_1}{2\sqrt{a_0a_2}}
$$

Lo primero que nos piden es un sistema subamortiguado, y seg√∫n el cuaderno de P√©ndulo El√°stico esto se da cuando el factor de amortiguamiento es:

$$
0<Œæ<1
$$

En este caso si $$a_0=1$$, $$a_1=1$$ y $$a_2=1$$ entonces:

$$
Œæ=\frac{1}{2\sqrt{1*1}}=\frac{1}{2}
$$

Y eso cumple con la condici√≥n, para esto $$m=1$$, $$c=1$$, $$k=1$$, $$R=1$$, $$L=1$$ y $$C=1$$.""")

st.markdown(r"""## Sistema subamortiguado

Ya vimos que para este caso

$$
Œæ=\frac{1}{2}
$$

Por su parte la frecuencia natural no amortiguada se calcula como:

$$
œâ_n=\sqrt{\frac{a_0}{a_2}}=\sqrt{\frac{1}{1}}=1
$$

Mientras que la frecuencia natural amortiguada es:

$$
œâ_d=\omega_n \sqrt{1-Œæ^2}=1*\sqrt{1-0,5^2}=\sqrt{0,75}‚âà0,866
$$

El tiempo pico es:

$$
t_p=\frac{œÄ}{œâ_n\sqrt{1-Œæ^2}}=\frac{œÄ}{1\sqrt{1-0,5^2}}‚âà3,628
$$

Y finalmente el tiempo de establecimiento es:

$$
t_s=\frac{3}{Œæœâ_n}=\frac{3}{0,5*1}=6
$$""")

st.markdown(r"""Ahora nos piden graficar los polos y ceros, para esto tenemos que hallar las raices de nuestra funci√≥n de transferencia:

$$
H(s)=\frac{1}{s^2+s+1}
$$

Como en el numerador no hay raices nuestra funci√≥n no tiene ceros, para hallar los polos usamos numpy:""")

import numpy as np

coef = [1, 1, 1]   # s^2 + s + 1
raices = np.roots(coef)

st.markdown(r"""Los polos son:""")

st.write(raices)


import matplotlib.pyplot as plt

fig12, ax12 = plt.subplots()

ax12.scatter(np.real(raices), np.imag(raices), marker='x', color='r', label='Polos')
ax12.axhline(0, color='black', linewidth=1) #ponemos los ejes para que se vea mejor
ax12.axvline(0, color='black', linewidth=1)
ax12.grid(True)
ax12.set_xlabel('Parte real')
ax12.set_ylabel('Parte imaginaria')
ax12.set_title('Plano s: Polos y Ceros')
ax12.legend(loc='upper right')


st.pyplot(fig12)

st.markdown(r"""Ahora vamos a hacer el diagrama de bode, para ello reemplazamos $$s$$ por $$jw$$ en la funci√≥n de transferencia y obtenemos:

$$
H(w)=\frac{1}{(jw)^2+jw+1}=\frac{1}{-w^2+jw+1}
$$

Ahora graficamos:""")

def H(w, m, c, k):
  return 1/(-(m*(w**2))+(1j*c*w)+k) #creamos nuestra funci√≥n de transferencia
w = np.logspace(-1,5,2000) #generamos un vector con valores igualmente espaciados en escala logaritmica
Heval = H(w, 1, 1, 1) #evaluamos la funci√≥n de transferencia en el vector w
Hmag = 20*np.log10(np.abs(Heval)) #magnitud de H en dB
Hphase = np.angle(Heval, deg=True) #fase de H

fig13, ax13 = plt.subplots()

ax13.semilogx(w, Hmag) #hace gr√°fico con eje x logaritmico
ax13.grid(True)
ax13.set_title("Magnitud (dB)")

st.pyplot(fig13)

fig14, ax14 = plt.subplots()

ax14.semilogx(w, Hphase)
ax14.grid(True)
ax14.set_title("Fase (grados)")

st.pyplot(fig14)

st.markdown(r"""Ahora tenemos que graficar la respuesta impulso. Para eso tenemos que solucionar la EDO, y usaremos Sympy para ello:""")

import sympy as sym

t, m, c, k = sym.symbols('t m c k', real=True)
x = sym.Function('x')(t)
y = sym.Function('y')(t)

ode = sym.Eq(m*y.diff(t, 2) + c*y.diff(t) + k*y, x)

st.latex(sym.latex(ode))

solution_h = sym.dsolve(ode.subs(x, sym.DiracDelta(t)).subs(y, sym.Function('h')(t)))

st.latex(sym.latex(solution_h))

st.markdown(r"""Determinamos constantes de integraci√≥n $$C1$$ y $$C2$$ a partir de condiciones iniciales $$y(t)=0$$ y $$\frac{dy(t)}{dt}=0$$ para $$t<0$$""")

integration_constants = sym.solve( (solution_h.rhs.limit(t, 0, '-'), solution_h.rhs.diff(t).limit(t, 0, '-')), ['C1', 'C2'] ) # sym.solve((dupla de expresiones de condiciones iniciales),(dupla de constantes))

st.latex(sym.latex(integration_constants))

st.markdown("Se sustituye:")

h = solution_h.subs(integration_constants)

st.latex(sym.latex(h))

st.markdown(r"""Ahora guardamos los valores de $$m$$, $$c$$ y $$k$$ en un  diccionario para sustituir y graficar:""")

mck = {m: 1, c: sym.Rational('1'), k: sym.Rational('1')}

st.latex(sym.latex(mck))

st.markdown(r"""Y graficamos:""")

hsub = h.subs(mck)

# Convertir a funci√≥n num√©rica
h_num = sym.lambdify(t, hsub.rhs, 'numpy')

# Rango de tiempo
ti = 0
tf = 12
N = 1000
T = np.linspace(ti, tf, N)

# Evaluar funci√≥n
H = h_num(T)

# Crear figura
figa, axa = plt.subplots()
axa.plot(T, H)
axa.set_xlabel("t")
axa.set_ylabel("h(t)")
axa.set_title("Respuesta h(t)")
axa.axhline(0, color='black', linewidth=1) #ponemos los ejes para que se vea mejor
axa.axvline(0, color='black', linewidth=1)
axa.grid(True)

# Mostrar en Streamlit
st.pyplot(figa)

st.markdown(r"""Ahora queremos graficar la respuesta escal√≥n, para eso nuevamente resolvemos la EDO por Sympy:""")

solution_u = sym.dsolve(ode.subs(x, sym.Heaviside(t)).subs(y, sym.Function('y_{step}')(t)))

st.latex(sym.latex(solution_u))

st.markdown("Constantes de integraci√≥n:")

integration_constants = sym.solve( (solution_u.rhs.limit(t, 0, '-'), solution_u.rhs.diff(t).limit(t, 0, '-')), ['C1', 'C2'] ) # sym.solve((dupla de expresiones de condiciones iniciales),(dupla de constantes))

st.latex(sym.latex(integration_constants))

ys = solution_u.subs(integration_constants)

st.latex(sym.latex(ys))

yssub = ys.subs(mck)

# Convertir a funci√≥n num√©rica
ys_num = sym.lambdify(t, yssub.rhs, 'numpy')

YS = ys_num(T)

# Crear figura
figb, axb = plt.subplots()
axb.plot(T, YS)
axb.set_xlabel("t")
axb.set_ylabel("y_{step}(t)")
axb.set_title("Respuesta step(t)")
axb.axhline(0, color='black', linewidth=1) #ponemos los ejes para que se vea mejor
axb.axvline(0, color='black', linewidth=1)
axb.grid(True)

# Mostrar en Streamlit
st.pyplot(figb)

st.markdown(r"""Ahora la respuesta a la rampa:""")

solution_r = sym.dsolve(ode.subs(x, t*sym.Heaviside(t)).subs(y, sym.Function('y_{ramp}')(t))) #la rampa es una funci√≥n lineal que empieza desde 0, por eso la definimos como t*escalon

st.latex(sym.latex(solution_r))

st.markdown("Constantes de integraci√≥n:")

yr= solution_r.rhs.subs(mck) #en esta ocasion haremos primero la sustituci√≥n ya que sino Sympy se pega

st.latex(sym.latex(yr))

integration_constants = sym.solve( (yr.limit(t, 0, '-'), yr.diff(t).limit(t, 0, '-')), ['C1', 'C2'] )

st.latex(sym.latex(integration_constants))

yrf = yr.subs(integration_constants)

st.latex(sym.latex(yrf))

# Convertir a funci√≥n num√©rica
yrf_num = sym.lambdify(t, yrf, 'numpy')

YRF = yrf_num(T)

# Crear figura
figc, axc = plt.subplots()
axc.plot(T, YRF)
axc.set_xlabel("t")
axc.set_ylabel("y_{ramp}(t)")
axc.set_title("Respuesta ramp(t)")
axc.axhline(0, color='black', linewidth=1) #ponemos los ejes para que se vea mejor
axc.axvline(0, color='black', linewidth=1)
axc.grid(True)

# Mostrar en Streamlit
st.pyplot(figc)

st.markdown(r"""### Sistema sobreamortiguado

Ya explicamos antes que para un buen m√©todo para buscar que un sistema cumpla con lo que queremos es por el factor de amortiguamiento $$Œæ$$ que viene dado por:

$$
Œæ=\frac{a_1}{2\sqrt{a_0a_2}}
$$

Con:

$$
H(s)=\frac{1}{a_2s^2+a_1s+a_0}
$$

Para que el sistema sea sobreamortiguado debe cumplirse que $$Œæ>1$$, con esto en mente podemos elegir $$a_2=1$$, $$a_1=3$$ y $$a_0=1$$, con lo que el factor de amortiguamiento ser√°:

$$
Œæ=\frac{a_1}{2\sqrt{a_0a_2}}=\frac{3}{2\sqrt{1*1}}=\frac{3}{2}
$$

Esto cumple con lo solicitado, asi $$m=1$$, $$c=3$$, $$k=1$$ y sus equivalentes en el circuito ser√°n $$L=1$$, $$R=\frac{1}{3}$$ y $$C=1$$

Una vez con el factor de amortiguamiento claro, la frecuencia natural no amortiguada es:

$$
œâ_n=\sqrt{\frac{a_0}{a_2}}=\sqrt{\frac{1}{1}}=1
$$

Mientras la frecuencia natural amortiguada es:

$$
œâ_d=œâ_n\sqrt{1-Œæ^2}=1*\sqrt{1-\left(\frac{3}{2}\right)^2} ‚âà j1,118
$$

Por su parte el tiempo pico es:

$$
t_p=\frac{œÄ}{œâ_n\sqrt{1-Œæ^2}}=\frac{œÄ}{1*\sqrt{1-\left(\frac{3}{2}\right)^2}} ‚âà -j2,81
$$

Y el tiempo de establecimiento es:

$$
t_s=\frac{3}{Œæœâ_n}=\frac{3}{\frac{3}{2}*1}=2
$$


Ahora vamos a graficar los polos y ceros. Para eso hallamos las raices de la ecuaci√≥n de transferencia:

$$
H(s)=\frac{1}{s^2+3s+1}
$$

""")

import numpy as np

coef2 = [1, 3, 1]   # s^2 + 3s + 1
raices2 = np.roots(coef2)

st.markdown(r"""Los polos son:""")

st.write(raices2)

figd, axd = plt.subplots()

axd.scatter(np.real(raices2), np.imag(raices2), marker='x', color='r', label='Polos')
axd.axhline(0, color='black', linewidth=1) #ponemos los ejes para que se vea mejor
axd.axvline(0, color='black', linewidth=1)
axd.set_xlabel('Parte real')
axd.set_ylabel('Parte imaginaria')
axd.set_title('Plano s: Polos y Ceros')
axd.legend(loc='upper right')
axd.grid(True)

st.pyplot(figd)

st.markdown(r"""Ahora para el diagrama de bode reemplazamos $$s$$ por $$jw$$ en la funci√≥n de transferencia:

$$
H(w)=\frac{1}{(jw)^2+3(jw)+1}=\frac{1}{-w^2+j3w+1}
$$""")

w = np.logspace(-1,5,2000) #generamos un vector con valores igualmente espaciados en escala logaritmica
Heval2 = H(w, 1, 3, 1) #evaluamos la funci√≥n de transferencia en el vector w
Hmag2 = 20*np.log10(np.abs(Heval2)) #magnitud de H en dB
Hphase2 = np.angle(Heval2, deg=True) #fase de H
fige, axe = plt.subplots()
axe.semilogx(w, Hmag2) #hace gr√°fico con eje x logaritmico
axe.grid(True)
axe.set_title("Magnitud (dB)")

figf, axf = plt.subplots()
axf.semilogx(w, Hphase2)
axf.grid(True)
axf.set_title("Fase (grados)")

st.pyplot(fige)
st.pyplot(figf)
```
